```python
from utils.objaverse_dataset import ObjaversePointCloudDataset

annotations_file = '/Users/kostyalbalint/Documents/Egyetem/7.Felev/Szakdolgozat/objaverse_labeling/concatenated_annotations.npy'
pc_dir = '/Users/kostyalbalint/Documents/Egyetem/7.Felev/Szakdolgozat/pointCloudsNoColor'

dataset = ObjaversePointCloudDataset(annotations_file=annotations_file, pc_dir=pc_dir, split='train', scale_mode='shape_unit', name_filter='human', load_to_mem=False)
```

```python
dataset.__len__()
```

```python
paths = []
for id in range(dataset.__len__()):
    paths.append(dataset.get_file_path(id).replace(pc_dir + "/", ""))
```

```python
paths
```

```python
import os
import shutil

destination_directory = '/Users/kostyalbalint/Documents/Egyetem/7.Felev/Szakdolgozat/pointCloudsSplit/human'

for file_path in paths:
    # Construct the full source and destination paths
    source_path = os.path.join(pc_dir, file_path)
    destination_path = os.path.join(destination_directory, file_path)

    # Ensure the destination directory exists, and copy the file
    os.makedirs(os.path.dirname(destination_path), exist_ok=True)
    shutil.copy2(source_path, destination_path)

print("Files have been copied to the destination directory.")
```
