```python
import torch

def sum_vector_lengths(tensor):    
    vector_lengths = torch.sqrt(torch.sum(tensor ** 2, dim=1))
    sum_lengths = torch.sum(vector_lengths)
    return sum_lengths
```

```python
lengths = []
for id in range(10):
    lengths.append(sum_vector_lengths(dataset[id]['pointcloud']))
```

```python
import plotly.graph_objects as go

def show_point_cloud(points, color, title):
    # Create a 3D scatter plot
    fig = go.Figure()

    # Scatter plot of the points
    fig.add_trace(go.Scatter3d(
        x=points[:, 0],
        y=points[:, 1],
        z=points[:, 2],
        mode='markers',
        marker=dict(
            size=2,
            color=color,  # You can customize the color here
        )
    ))

    # Set layout options
    fig.update_layout(scene=dict(
        xaxis_title='X-axis',
        yaxis_title='Y-axis',
        zaxis_title='Z-axis'),
        title=title)

    # Show the plot
    return fig.show()

show_point_cloud(dataset[4]['pointcloud'], 'red', '')
```

```python
paths = []
for id in range(dataset.__len__()):
    paths.append(dataset.get_file_path(id).replace(pc_dir + "/", ""))
```

```python
paths
```

```python
import os
import shutil

destination_directory = '/Users/kostyalbalint/Documents/Egyetem/7.Felev/Szakdolgozat/pointCloudsSplit/knife'

for file_path in paths:
    # Construct the full source and destination paths
    source_path = os.path.join(pc_dir, file_path)
    destination_path = os.path.join(destination_directory, file_path)

    # Ensure the destination directory exists, and copy the file
    os.makedirs(os.path.dirname(destination_path), exist_ok=True)
    shutil.copy2(source_path, destination_path)

print("Files have been copied to the destination directory.")
```
